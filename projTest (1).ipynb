```python
df.dropna(subset=['loc'],inplace=True)    
df.dropna(subset=['Engine Type'],inplace=True)  
df.dropna(subset=['Turbo Charger'],inplace=True) 
df.dropna(subset=['Super Charger'],inplace=True) 
df.dropna(subset=['Drive Type'],inplace=True) 
df.dropna(subset=['Gear Box'],inplace=True) 
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Function to clean string columns
def clean_column(col, patterns):
    col = col.astype(str)
    for pat in patterns:
        col = col.str.replace(pat, '', regex=True)
    return col.str.strip()

# Dictionary of cleaning patterns
patterns = {
    'Length': [',', 'mm'],
    'Width': [',', 'mm'],
    'Height': [',', 'mm'],
    'Wheel Base': [',', 'mm', 'm'],
    'Acceleration': ['Seconds', 'seconds', 's', 'ec', 'S', 'ond', 'c', '.', 'm'],
    'Cargo Volumn': [r'\D', '-', 'litre', 'liter', 'L', r'\(3rd Row Foded\)', 's'],
    'mileage_new': ['kmpl', 'km/kg'],
    'Gross Weight': ['kg', 'Kg', 's', ',', '.'],
    'Top Speed': ['kmph', 'Kmph', 'km/hr', 'Km/hr', 'Km/h', 'km/h', 'Km', r'\(claimed\)'],
    'Kerb Weight': ['kg', 'Kg', 's', ',', '.', 'mm'],
}

# Apply cleaning and convert to numeric
for col, pats in patterns.items():
    df[col] = clean_column(df[col], pats)
    if col == 'Kerb Weight':
        df[col] = df[col].apply(lambda x: (sum(map(float, x.split('-'))) / 2) if '-' in x else x)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop columns that are fully empty
df = df.dropna(axis=1, how='all')

# Check if 'Acceleration' exists in the DataFrame
if 'Acceleration' in df.columns:
    # Impute numeric columns (only if column has at least 1 non-NaN)
    imputer = SimpleImputer(strategy='mean')
    for col in list(patterns.keys()) + ['max_engine_capacity_new', 'Values per Cylinder', 'No of Cylinder']:
        if col in df.columns and df[col].notna().sum() > 0:
            df[col] = imputer.fit_transform(df[[col]]).ravel()

    # Gear Box: Extract first digit
    df['Gear Box'] = df['Gear Box'].astype(str).str.extract(r'(\d)').astype(float)

    # Only impute if not fully empty
    if df['Gear Box'].notna().sum() > 0:
        df['Gear Box'] = imputer.fit_transform(df[['Gear Box']]).ravel()

    # Final structure check
    print(df.info())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer


X_test = df.drop("dynx_totalvalue_y", axis=1)
y_test = df["dynx_totalvalue_y"]

# Define column types
numerical_cols = X_test.select_dtypes(include=["float64", "int64"]).columns
categorical_cols = X_test.select_dtypes(include=["object"]).columns

# Limit categories to top 5
def limit_top_5_categories(data, categorical_cols):
    for col in categorical_cols:
        top_5 = data[col].value_counts().nlargest(5).index
        data[col] = data[col].apply(lambda x: x if x in top_5 else 'Other')
    return data

# Merge similar encoded columns safely
def merge_similar_columns(df, merge_dict):
    for new_col, col_group in merge_dict.items():
        available_cols = [col for col in col_group if col in df.columns]
        if available_cols:
            df[new_col] = df[available_cols].max(axis=1)
            df.drop(columns=available_cols, inplace=True)
    return df

# Add binary price flag for high prices
def add_price_flags(df, price_series):
    df['is_high_price'] = (price_series > price_series.quantile(0.90)).astype(int)
    return df

# Drop specific encoded features
def drop_unwanted_encoded_features(df):
    columns_to_drop = [col for col in df.columns if "model_type_new_used" in col or "seller_type_new_dealer" in col]
    df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
    return df

# Preprocess validation and test sets
def preprocessing_other(data, y_data, scaler, one_hot_encoder, imputer):
    # Filter numerical columns only
    numerical_cols_clean = data[numerical_cols].select_dtypes(include=["float64", "int64"]).columns

    # Impute missing values on numerical columns
    data[numerical_cols_clean] = imputer.transform(data[numerical_cols_clean])
    
    # Scale numerical columns
    data[numerical_cols_clean] = scaler.transform(data[numerical_cols_clean])

    # Limit categorical values
    data = limit_top_5_categories(data, categorical_cols)

    # One-hot encode categorical columns
    encoded_categorical = one_hot_encoder.transform(data[categorical_cols])
    encoded_df = pd.DataFrame(
        encoded_categorical,
        columns=one_hot_encoder.get_feature_names_out(categorical_cols),
        index=data.index
    )

    # Merge similar encoded columns
    column_merges = {
        'Super Charger_Yes': ['Super Charger_Yes', 'Super Charger_yes'],
        'Super Charger_No': ['Super Charger_No', 'Super Charger_no'],
        'Turbo Charger_Yes': ['Turbo Charger_Yes', 'Turbo Charger_yes', 'Turbo Charger_Twin'],
    }
    encoded_df = merge_similar_columns(encoded_df, column_merges)

    # Drop unwanted encoded features
    encoded_df = drop_unwanted_encoded_features(encoded_df)

    # Combine the transformed numerical and encoded features
    df_encoded = pd.concat([data.drop(columns=categorical_cols), encoded_df], axis=1)
    
    # Add high-price flag using y_data (for test data, using y_test)
    df_encoded = add_price_flags(df_encoded, y_data)

    return df_encoded, np.log1p(y_data)

# Preprocess test data (match structure with validation and training)
X_test_transformed, y_test_log = preprocessing_other(X_test.copy(), y_test, scaler, one_hot_encoder, imputer)
```

```python
predictions = model4.predict(X_test_transformed)
```
